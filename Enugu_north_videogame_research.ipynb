From the output above, it is evident the Age format is formatted as a datetime datatype. Although there's something quite similar about all the entries and it's the fact that they all begin with '2025-'. We could easily replace all instances of '2025-' with nothing.
```python
# Replace '2025-' with an empty space.
df['Age'] = df['Age'].astype(str)
df['Age'] = df['Age'].str.replace('2025-', '')
df['Age']
```

There seems to be no significant missing values. Just the second entry which is negligible in this analysis. Next up, I'll encode the data. in this case we'd be making use of ordinal encoding. I will assign a number to each of the ordinals

There are 3 sets of four Ordinal data:

(Students' Questionnaire) **SECTION A**
*   More than 5hours daily = 4
*   2-5hours daily = 3

*  Less than 2hours daily = 2
*   Never = 1



(Parents' Questionnaire) **SECTION B**
*   Strongly agree = 4

*   Agree = 3
*   Disagree = 2


*   Strongly disagree = 1


(Screen time control) **SECTION C**



  Sometimes

*   Always = 4
*   Often = 3
*   Sometimes = 2
*   Never = 1









```python
# Perform ordinal encoding on the entire dataset based on the predefined values
ordinal_values = {
    'more than 5hrs daily': 4,
    '2-5hrs daily': 3,
    'less than 2hrs daily': 2,
    'never': 1,
    'strongly agree': 4,
    'agree': 3,
    'disagree': 2,
    'strongly disagree': 1,
    'always': 4,
    'often': 3,
    'sometimes': 2,
    'never': 1
    }

# Apply the mapping to the entire DataFrame
df = df.applymap(lambda x: ordinal_values.get(x, x))  # Replace only if the value exists in the mapping
df.head()
```

This looks good, however, we may need to add another column to help with

```python
# List of game columns
games_list = ['Tetris', 'Plant_vs.Zombies', 'AngryBird', 'Two_zero_four_eight', 'Subway_Surfers', 'Fruit_Ninja', 'Temple_Run', 'Minecraft', 'Fortenite', 'Mario_Kart_Deluxe',
            'Genshin_Impact', 'Animal_Crossing', 'Roblox', 'Clashes_of_Clans', 'Final_Fantasy_XV', 'Raid_shawdow_legends', 'AFK_Arena', 'The_Elder_Scrolls', 'Pokemon_GO',
            'Eight_Ball_Pool', 'Call_of_Duty', 'Brawl_Stars', 'Clash_Royale', 'King_of_Avalon', 'Rise_of_Kingdoms']

# Initialize a dictionary to store the frequency counts
frequency_counts = {
    'More than 5 hours daily': 0,
    '2-5 hours daily': 0,
    'Less than 2 hours daily': 0,
    'Never': 0
}

# Iterate through each game column and count the frequency of each category
for game in games_list:
    frequency_counts['More than 5 hours daily'] += (df[game] == 4).sum()
    frequency_counts['2-5 hours daily'] += (df[game] == 3).sum()
    frequency_counts['Less than 2 hours daily'] += (df[game] == 2).sum()
    frequency_counts['Never'] += (df[game] == 1).sum()

# Convert the counts to percentages
total_responses = len(df) * len(games_list)  # Total possible responses
frequency_percentages = {k: (v / total_responses) * 100 for k, v in frequency_counts.items()}

# Display the results
print("Frequency of Playing Video Games:")
for category, percentage in frequency_percentages.items():
    print(f"{category}: {percentage:.2f}%")
```

```python
# Data for visualization
categories = list(frequency_percentages.keys())
percentages = list(frequency_percentages.values())

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(categories, percentages, color=['brown'])
plt.xlabel('Frequency of Playing Video Games')
plt.ylabel('Percentage of Responses (%)')
plt.title('Frequency of Playing Video Games Among Children')
plt.ylim(0, 100)  # Set the y-axis limit to 0-100%
plt.show()
```

Now I will need to Answer the second question which is: **What are parentsâ€™ attitudes towards video game as an educational tool?**

The following column labels; 'QB1', 'QB2', 'QB3', 'QB4', 'QB5', 'QB6', 'QB7', 'QB8', 'QB9', 'QB10', 'QB11', 'QB12' and 'B13'  are aliases, you can find their questions they represent in the Readme.

```python
# Create a list of all required columns
parent_att_ques = ['QB1', 'QB2', 'QB3', 'QB4', 'QB5', 'QB6', 'QB7', 'QB8', 'QB9', 'QB10', 'QB11', 'QB12', 'QB13']

# Calculate the mean value of each column (Section B)
mean_scores = df[parent_att_ques].mean()

# Print result
print("Parents' attitudes to video games:")
print(mean_scores)
```

```python
# Data for visualization
mean_scores_sorted = mean_scores.sort_values(ascending=False)
questions = mean_scores_sorted.index
scores = mean_scores_sorted.values

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.barh(questions, scores, color='brown')
plt.xlabel('Mean Score (1 = Strongly Disagree, 4 = Strongly Agree)')
plt.title("Parents' Attitudes Towards Video Games as an Educational Tool")
plt.xlim(1, 4)  # Set the x-axis limits to match the Likert scale
plt.show()
```

Now I will answer the very last question: **What are the strategies used by parents in monitoring their childrenâ€™s screen time?**

```python
# Create a list of all required columns
parent_strat_ques = ['QC1', 'QC2', 'QC3', 'QC4', 'QC5', 'QC6', 'QC7', 'QC8', 'QC9', 'QC10', 'QC11', 'QC12', 'QC13']

# Calculate the mean value of each column (Section C)
mean_strat_scores = df[parent_strat_ques].mean()

# Print result
print("Parents's screentime monitoring strategies:")
print(mean_strat_scores)
```

```python
# Sort the mean scores in descending order
mean_strat_scores_sorted = mean_strat_scores.sort_values(ascending=False)

# Extract the sorted strategies and scores
strategies_sorted = mean_strat_scores_sorted.index
scores_sorted = mean_strat_scores_sorted.values

# Create a bar chart
plt.figure(figsize=(12, 8))
plt.barh(strategies_sorted, scores_sorted, color='brown')
plt.xlabel('Mean Score (1 = Never, 4 = Always)')
plt.title("Strategies Used by Parents to Monitor Children's Screen Time")
plt.xlim(1, 4)  # Set the x-axis limits to match the response scale
plt.show()
```

```python
# Initialize a dictionary to store the frequency of agreement for each strategy
strategy_frequencies = {}

# Calculate the frequency of agreement (Strongly Agree + Agree) for each strategy
for strategy in parent_strat_ques:
    agreement_count = df[(df[strategy] == 4) | (df[strategy] == 3)].shape[0]
    strategy_frequencies[strategy] = agreement_count

# Convert the counts to percentages
total_parents = df.shape[0]
strategy_percentages = {k: (v / total_parents) * 100 for k, v in strategy_frequencies.items()}

# Display the results
print("Frequency of Parents Using Each Strategy:")
for strategy, percentage in strategy_percentages.items():
    print(f"{strategy}: {percentage:.2f}%")
```

```python
# Map numerical values to frequency categories
value_to_category = {
    4: 'More than 5 hours daily',
    3: '2-5 hours daily',
    2: 'Less than 2 hours daily',
    1: 'Never'
}

# Convert the game columns to numeric, forcing non-numeric values to NaN
for game in games_list:
    df[game] = pd.to_numeric(df[game], errors='coerce')

# Now proceed with calculating Video_Game_Frequency
df['Video_Game_Frequency'] = (
    df[games_list]              # Select all game columns
    .max(axis=1)                # Find the highest value (4, 3, 2, or 1) for each student
    .map(value_to_category)     # Map the value to its corresponding category
)

# Display the DataFrame with the new column
print("DataFrame with Video_Game_Frequency:")
print(df[['First_Term_2023.2024(1)', 'Second_Term_2023.2024(2)', 'Third_Term_2024.2025(3)', 'Video_Game_Frequency']].head())
```
